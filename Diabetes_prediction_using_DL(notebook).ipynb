Step1: Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

STEP2: READING the data

```python
df=pd.read_csv('/content/diabetes.csv')
```

Step 4:

```python
df.Outcome.value_counts()
```

```python
x=df.drop('Outcome',axis=1)
```

```python
x
```

```python
y=df.Outcome
```

```python
y
```

Step5:

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=30)
```

Step6:

```python
Lr=LogisticRegression()
Lr.fit(xtrain,ytrain)
```

```python
ypred=Lr.predict(xtest)
```


Step7:

```python
LRacc=accuracy_score(ytest,ypred)
LRacc
```

```python
DT=DecisionTreeClassifier()
DT.fit(xtrain,ytrain)
```

```python
DTpred=DT.predict(xtest)
```

```python
DTacc=accuracy_score(ytest,DTpred)
DTacc
```

```python
RF=RandomForestClassifier()
RF.fit(xtrain,ytrain)
```

```python
RFpred=RF.predict(xtest)
RFacc=accuracy_score(ytest,RFpred)
RFacc
```

```python
import joblib
joblib.dump(Lr,'LogM.pkl')
joblib.dump(DT,'DT.pkl')
joblib.dump(RF,'RF.pkl')
lr=joblib.load('LogM.pkl')
lr
```

```python
import numpy as np
import gradio as gr
```

```python
def predictD(Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age):
  q=np.array([Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age]).reshape(-1,1)
  lr=joblib.load('LogM.pkl')
  dtm=joblib.load('DT.pkl')
  rfm=joblib.load('RF.pkl')
  lrma=rfm.predict(q)
  if(int(lrma)==1):
    return'The patient is diabitics.'

  else:
    return 'The patient is not diabitics.'
```

```python
import gradio as gr
iface=gr.Interface(fn=predictD,
                      inputs=[gr.Number(label='Enter the umber of pregnancies:'),
                              gr.Number(label='Enter the glucose level:'),
                              gr.Number(label='Enter the BP:'),
                              gr.Number(label='Enter the skinthickness:'),
                              gr.Number(label='Enter the isuline:'),
                              gr.Number(label='Enter the BMI:'),
                              gr.Number(label='Enter the DiabetesPedigreeFunction:'),
                              gr.Number(label='Enter the Age:')],
                      outputs='text')
```
